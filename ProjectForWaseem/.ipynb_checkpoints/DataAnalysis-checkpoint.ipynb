```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.rcParams['patch.force_edgecolor'] = True
import seaborn as sns
%matplotlib inline
import json
```

```python
# Loading dataset and see the columns
dataset = pd.read_csv('data_analyst_rec_assignment_data.csv')
dataset.columns
```

Check if the data is clean

```python
sns.heatmap(dataset.isnull(), yticklabels=False, cbar=False, cmap = 'viridis')
```

There are a few NaN in the estimated_country column, but that should bother us from doing the basic analysis

### Plot the distribution of the install and click count for different campaigns

```python
plt.figure(figsize=(15, 7))
sns.set_context('poster',font_scale=1)
sns.boxplot(x = 'campaign_id', y = 'install_count', data = dataset, palette= 'winter')
```

```python
plt.figure(figsize=(15, 7))
sns.boxplot(x = 'campaign_id', y = 'click_count', data = dataset, palette= 'winter')
```

# Grouping by campaign_id

Since the campaigns are run on different days, countries, etc., it makes sense to compare metrics like install_count and click_count by grouping all the campaign_ids together. 

```python
# We take the sum of all the metrics belonging to a particular campaign id
dataset_grouped = dataset.groupby('campaign_id').sum()
dataset_grouped

## Note: you can extend the analysis by grouping many other metrics like campaign_id 
#  and influencer_id etc. 
## dataset_grouped = dataset.groupby(['campaign_id','influencer_id']).sum()
```

```python
# Finding corelation between different metrics
corrmat = dataset_grouped.corr()
f, ax = plt.subplots(figsize=(12, 9))
sns.heatmap(corrmat, vmax=1.0, square=True);

# Priting the corelation matrix as well 
print(corrmat)
```

We find a strong corelation between install_count and subscriber_count and a some-what strong corelation between install_count and video_count/view_count.  

We find a some-what strong corelation between click_count and subscriber_count, video_count and view_count. 

There exists a strong corelation between subscriber_count, video_count and view_count (expected!!!)

We confirm this in the following pair plots 

```python
cols = ['install_count',
       'click_count',
       'subscriber_count', 'video_count', 'view_count',
       ]
sns.pairplot(dataset_grouped[cols], size = 2.2)
plt.show();
```

Plotting all the linear relationships

```python
sns.set()
fig, axes = plt.subplots(nrows=1, ncols=3,figsize=(18,5))
# Subcriber_count shows very strong linear relationship, make sure you highlight this in the report!!!! 
sns.regplot(x = 'subscriber_count', y = 'install_count', data = dataset_grouped, ax = axes[0])
sns.regplot(x = 'view_count', y = 'install_count', data = dataset_grouped, ax = axes[1])
## Check the third plot, not sure why it isn't coming cleanly here
sns.regplot(x = 'video_count', y = 'install_count', data = dataset_grouped, ax = axes[2])
```

```python
# alternative for the 3rd plot above
sns.lmplot(x = 'video_count', y = 'install_count', data = dataset_grouped)
```

```python
sns.set()
fig, axes = plt.subplots(nrows=1, ncols=3,figsize=(18,5))
# Strong linear relationship between subscriber_count and click_count
sns.regplot(x = 'subscriber_count', y = 'click_count', data = dataset_grouped, ax = axes[0])
sns.regplot(x = 'view_count', y = 'click_count', data = dataset_grouped, ax = axes[1])
## Check the third plot, not sure why it isn't coming cleanly here
sns.regplot(x = 'video_count', y = 'click_count', data = dataset_grouped, ax = axes[2])
```

```python
# alternative for the 3rd plot above
sns.lmplot(x = 'video_count', y = 'click_count', data = dataset_grouped)
```
